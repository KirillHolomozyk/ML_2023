```python
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt

# Завантаження даних MNIST
mnist = tf.keras.datasets.mnist
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
```

```python
Створення нейронної мережі:
Розробіть архітектуру нейронної мережі, наприклад:
```

```python
# Нормалізація значень пікселів до діапазону [0, 1]
train_images, test_images = train_images / 255.0, test_images / 255.0
```

```python
model = models.Sequential()
model.add(layers.Flatten(input_shape=(28, 28)))  # Вхідний шар
model.add(layers.Dense(128, activation='relu'))   # Повнозв'язний шар 1
model.add(layers.Dense(10, activation='softmax'))  # Повнозв'язний шар 2 (вихідний)
```

```python
Тренування моделі:
Використовуйте оптимізатор, функцію втрат та метрики для тренування моделі:
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(train_images, train_labels, epochs=5)
```

```python
Оцінка ефективності моделі:
Використовуйте тестовий набір для оцінки точності моделі:
```

```python
test_loss, test_acc = model.evaluate(test_images, test_labels)
print(f'Test accuracy: {test_acc}')
```

```python
Візуалізація результатів:
Виведіть деякі приклади зображень та їх класифікацію:
```

```python
predictions = model.predict(test_images)

# Вибір випадкового зображення для відображення
for i in range(10):
    plt.imshow(test_images[i], cmap='gray')
    plt.title(f'Actual: {test_labels[i]}, Predicted: {tf.argmax(predictions[i])}')
    plt.show()

```
