```python
import numpy as np

# Створення набору даних
size = 10
scalar = 5
vector = np.arange(size)
matrix = np.random.rand(size, size)

# CRUD операції
vector += scalar
matrix *= scalar

print("Scalar:", scalar)
print("Vector:", vector)
print("Matrix:", matrix)
```

```python
import pandas as pd

# Створення DataFrame
data = {'Name': ['John', 'Jane', 'Bob'],
        'Age': [25, 30, 22],
        'Salary': [50000, 60000, 45000]}

df = pd.DataFrame(data)

# Маніпуляції з даними
print(df.head())
print(df.describe())
print(df.iloc[0])  # вибір по індексу
print(df.loc[df['Age'] > 25])  # вибір за умовою
```

```python
import matplotlib.pyplot as plt

# Створення набору даних лінійної функції з помилками
X = np.arange(0, 10, 0.1)
Y_true = 2 * X + 1
noise = np.random.normal(0, 1, len(X))
Y_noisy = Y_true + noise

# Побудова графіку
plt.scatter(X, Y_noisy, label='Noisy Data')
plt.plot(X, Y_true, label='True Function', color='red')
plt.legend()

# Обчислення похибок
MAE = np.mean(np.abs(Y_true - Y_noisy))
MSE = np.mean((Y_true - Y_noisy)**2)

# Запис результатів у CSV файл
results = pd.DataFrame({'X': X, 'Y_true': Y_true, 'Y_noisy': Y_noisy})
results.to_csv('results.csv', index=False)

print("MAE:", MAE)
print("MSE:", MSE)
```
